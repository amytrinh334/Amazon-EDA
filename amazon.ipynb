Step 1: Loading Dataset

```python
#Importing required modules
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
#Loading the dataset
amazon = pd.read_csv("amazon.csv")
```

Step 2: Data Cleaning

```python
amazon.isnull().sum()
```

```python
amazon.dropna()
```

```python
#Converting certain columns to numerical data types for later use
amazon['discounted_price'] = amazon['discounted_price'].str.replace('₹', '', regex=False).str.replace(',', '', regex=False)
amazon['actual_price'] = amazon['actual_price'].str.replace('₹', '', regex=False).str.replace(',', '', regex=False)
amazon['discount_percentage'] = amazon['discount_percentage'].str.replace('%', '', regex=False)
amazon['rating_count'] = amazon['rating_count'].str.replace(',', '', regex=False)

columns = ["discounted_price","actual_price","discount_percentage","rating_count"]
for col in columns:
    amazon[col]=amazon[col].astype(float)
```

```python
amazon.head()
```

```python
#converting Rupees into USD
rupees = ["discounted_price","actual_price"]
for col in rupees:
    amazon[col] = amazon[col] * 0.011
    amazon[col]= amazon[col].round(2)
```

```python
amazon.head()
```

```python
#Simplifying category names
split_category = amazon['category'].str.split('|')
amazon['category'] = split_category.str[0]
amazon['item_type'] = split_category.str[-1]

amazon['category'] = amazon['category'].str.replace('&', ' & ', regex=False)
amazon['item_type'] = amazon['item_type'].str.replace('&', ' & ', regex=False)

amazon.head()
```

Step 3: Data Analysis

```python
#Barplot for ratings
amazon['rating'] = pd.to_numeric(amazon['rating'], errors='coerce')

plt.figure(figsize=(10,6))
sns.countplot(data=amazon, x="rating", palette="Wistia")
plt.title("Frequency of Product Ratings")
plt.xlabel("Rating")
plt.ylabel("Counts")
plt.show()
```

```python
#numerical variable heatmap
plt.figure(figsize=(10,6))
correlation = amazon[["discounted_price","actual_price","discount_percentage","rating","rating_count"]]
sns.heatmap(correlation.corr(), annot=True, cmap="vlag")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=amazon, x='category', y='rating', inner='box')
plt.xticks(rotation=45)
plt.title('Rating Density per Category')
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(
    data=amazon, 
    y='category', 
    order=amazon['category'].value_counts().index,
    palette='viridis'
)

plt.title('Most Popular Project Categories', fontsize=15)
plt.xlabel('Number of Products')
plt.ylabel('Category')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

top_10 = amazon['item_type'].value_counts().head(10).index


sns.countplot(
    data=amazon[amazon['item_type'].isin(top_10)], 
    y='item_type', 
    order=top_10,
    palette='viridis'
)

plt.title('Top 10 Most Popular Product Types', fontsize=15)
plt.xlabel('Number of Products')
plt.ylabel('Item Type')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(
    data=amazon, 
    x='discounted_price', 
    y='rating', 
    alpha=0.4,
    color='darkblue'
)
plt.title('Discount Price vs. Product Rating', fontsize=15)
plt.xlabel('Discount Price (USD)', fontsize=12)
plt.ylabel('Rating', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(
    data=amazon, 
    x='discount_percentage', 
    y='rating', 
    alpha=0.4,
    color='darkblue'
)
plt.title('Discount Percentage vs. Product Rating', fontsize=15)
plt.xlabel('Discount Percentage (%)', fontsize=12)
plt.ylabel('Rating', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
